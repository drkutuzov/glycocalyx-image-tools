```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
from scipy.special import erf
```

```python
def gcx(x, y, s, d):
    x1, x2 = (x - d/2)/np.sqrt(2)/s, (x + d/2)/np.sqrt(2)/s
    return 0.5*(1 + erf(x1) - erf(x2))*np.exp(-y**2/2/s**2)
```

```python
x, y = np.linspace(-10, 10, 128), np.linspace(-10, 10, 128)
X, Y = np.meshgrid(x, y)
```

```python
image = gcx(X, Y, 1, 6)
plt.imshow(image, extent=(x[0], x[-1], y[0], y[-1]), cmap='afmhot')
```
