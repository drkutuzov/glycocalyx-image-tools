```python
import matplotlib.pyplot as plt
import numpy as np
import gcx_map
```

## Line parameters: 

```python
point1, point2 = (-6, 3), (2, -7) 
center = (0.5*(point1[0] + point2[0]), 0.5*(point1[1] + point2[1]))
```

## Image grid

```python
x, y = np.linspace(-10, 10, 128), np.linspace(-10, 10, 128)
X, Y = np.meshgrid(x, y)
```

## Plot glycocalyx segment

```python
im = plt.imshow(gcx_map.gcx_segment(X, Y, 10, 1, 6, *center, gcx_map.get_angle(point1, point2)), 
                
                extent=(x[0], x[-1], y[0], y[-1]), 
                cmap='afmhot', 
                origin='lower')

plt.plot((point1[0], point2[0]), (point1[1], point2[1]), ls='--', lw=1.5, c='m')

cbar = plt.colorbar(im)
```
